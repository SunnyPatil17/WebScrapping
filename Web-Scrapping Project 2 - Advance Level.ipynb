```python
# Imports

from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

```python
# Store website as variable

website = 'https://www.yellowpages.com/search?search_terms=restuarant&geo_location_terms=New+York%2C+NY'
```

```python
response = requests.get(website)
response.status_code
```

```python
soup = bs(response.content, 'html.parser')
soup
```

```python
result_container = soup.find_all('div', {'class':'result'})
len(result_container)
```

```python
# we combine url part 1 with url part 2 in order to get absolute url

url_part1 = 'https://www.yellowpages.com/'

url_part2 = []
#loop through results and then through links

for item in result_container :
    for link in item.find_all('a', {'class':'business-name'}) :
        url_part2.append(link.get('href'))
        
```

```python
import urllib.parse

url_joined = []

for link in url_part2 :
    url_joined.append(urllib.parse.urljoin(url_part1,link))

url_joined
```

```python
first_link = url_joined[0]
first_link
```

```python
response = requests.get(first_link)
response.status_code
```

```python
soup = bs(response.content, 'html.parser')
soup
```

```python
# Extract Name
soup.find('h1',{'class':'dockable business-name'}).get_text().replace('\ufeffAdd to Favorites','')
```

```python
# Extract Phone Number
soup.find('a', {'class':'phone dockable'}).get_text()
```

```python
# Extract Address
soup.find('a',{'class':'directions small-btn'}).get_text().replace('Map & Directions','')
```

```python
soup.find('dd', {'class':'general-info'})
```

```python
# empty list - we did it so we can append all results - hand it over to pandas dataframe
results = []

# loop through all links
for link in url_joined :
    response = requests.get(link)
    soup = bs(response.content, 'html.parser')
    try:
        name = soup.find('h1',{'class':'dockable business-name'}).get_text().replace('\ufeffAdd to Favorites','')
    except:
        name = 'n/a'
        
    try:
        address = soup.find('a',{'class':'directions small-btn'}).get_text().replace('Map & Directions','')
    except:
        address = 'n/a'
    
    try:
        phone_num = soup.find('a', {'class':'phone dockable'}).get_text()
    except:
        phone_num = 'n/a'
    
    try:
        general_info = soup.find('dd', {'class':'general-info'}).get_text().replace('"','')
    except:
        general_info = 'n/a'
        
    # create dictinory with result
    
    output = {'Resturant Name': name , 'Address': address, 'Phone Number': phone_num, 'General Info': general_info}
    
    
    # append results in empty list
    results.append(output)
    
```

```python
df = pd.DataFrame(results)
df
```

```python
df.to_excel('project2_result.xlsx', index=False)
```
